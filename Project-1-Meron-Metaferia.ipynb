**2.a. Selecting Specific Years and Countries** 

```python
#A function using column selection method 

def year(df):
    
    ''' Function to select years 2011-2019 
        for the dataframes that were imported above'''
    
    Selected = df[['Country Name', 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019]]
    return Selected
```

```python
#applying the function above to each of the dfs

G_abs_y = year(G_tot_dol)
#df_GLCU = year(G_tot_LCU)
G_perCap_y = year(G_pC_dol)
#df_GpCLCU = year(G_pC_LCU)
M_abs_y = year(M_tot_dol)
#df_MLCU = year(M_tot_LCU)
M_perCap_y = year(M_PCap)
share_GDP_y = year(share_GDP)
Pop_y = year(Pop)
```

```python
#checking if selection has been applied

share_GDP_y.head() 
```

```python
#isin() conditional function to select the 10 countries
#sort alphabetically by country name for ease of joining the df later

def countries(df):
    
    '''Function to select the 10 countries that are 
        listed under Introduction and do sort by Country Name.'''
    
    Selected = df[df["Country Name"].isin(['China', 'Russia', 'Germany', 'United Kingdom', 'France', 'Italy', 'Iran', 'Saudi Arabia', 'Korea, South', 'Israel'])]
    Selected.sort_values(by = 'Country Name', inplace=True)
    return Selected
```

```python
#applying the function above to each dfs
G_abs = countries(G_abs_y)
G_Pcap = countries(G_perCap_y)
M_abs = countries(M_abs_y)
M_Pcap = countries(M_perCap_y)
Pop_d = countries(Pop_y)
M_share_G = countries(share_GDP_y)
```

**2.c. Rearranging Df for Ease of Analysis and Joining**

```python
#using melt function to format df where country name will be the identifier variable and the dollar amount, %, Pop. data on a given year will be the measured values

g_abs = G_abs.melt(['Country Name'], var_name = 'Year', value_name = 'GDP_abs')
g_Pcap = G_Pcap.melt(['Country Name'], var_name = 'Year', value_name = 'GDP_Per_Capita')
m_abs = M_abs.melt(['Country Name'], var_name = 'Year', value_name = 'Military_Expenditure')
m_Pcap= M_Pcap.melt(['Country Name'], var_name = 'Year', value_name = 'Per_Capita_Military_Expenditure')
m_P_g = M_share_G.melt(['Country Name'], var_name = 'Year', value_name = 'Share_of_GDP')
POP = Pop_d.melt(['Country Name'], var_name = 'Year', value_name = 'Population')

POP
```

**2.d.Converting GDP_abs and Military_abs values to correct USD dollar Values**

```python
#From observing the G_abs and M_abs value we see that the value amounts in dollars are not in consistent number units
# GDP_abs --> convert the values to trillion by dividing the GDP_abs series by (10^12)
#pandas series.apply function for conversion on each row value

g_abs['GDP_Absolute(USD trillions)'] = g_abs['GDP_abs'].apply(lambda x: x/1_000_000_000_000)
g_abs
```

```python
#dropping the GDP_abs column for ease of joining 

g_ab = g_abs.drop(['GDP_abs'], axis = 1) 
g_ab
```

```python
#since our Sipiri military data mentions given values are in millions for the abs_military 
## military_abs USD --> divide Military_Expenditure series by 1000 to convert to billions

m_abs['Military_Expenditure(USD billions)'] = m_abs['Military_Expenditure'].apply(lambda x: x/1000)
m_abs
```

```python
#dropping the Military_Expenditure column for ease of joining 

m_ab = m_abs.drop(['Military_Expenditure'], axis = 1) 
m_ab
```

```python
#change share of GDP to percentages

m_P_g['Share of GDP(%)'] = m_P_g['Share_of_GDP'].apply(lambda x: x*100)
m_P_g
```

```python
#dropping Share_of_GDP column for ease of joining

mPg = m_P_g.drop(['Share_of_GDP'], axis = 1) 
mPg
```

```python
# To be able to see the whole df after I joined and sort to check if my cleaning has been applied correctly

pd.set_option("display.max_rows", None, "display.max_columns", None)
```

**2.e. Merging DataFrames**

```python
#merging abs_GDP and PerCapita_GDP

g_merged = pd.merge(g_ab, g_Pcap, how='left', left_on=['Country Name', 'Year'], right_on=['Country Name', 'Year'], suffixes = ['GDP_Absolute(USD trillions)','GDP_Per_Capita'])
g_merged
```

```python
#merging abs_Mil_Exp. and PerCapita_Mil_Expe

m_merged =  pd.merge(m_ab, m_Pcap, how='left', left_on=['Country Name', 'Year'], right_on=['Country Name', 'Year'], suffixes = ['Military_Expenditure(USD billions)','Per_Capita_Military_Expenditure'])
m_merged
```

```python
#merging share_of_GDP and Pop. df

mPg_Pop = pd.merge(mPg, POP, how='left', left_on=['Country Name', 'Year'], right_on=['Country Name', 'Year'], suffixes = ['Share of GDP(%)','Population'])
mPg_Pop
```

```python
#merging the 3 dfs above to give one big merged df that has all the information together

merged_full = pd.merge(pd.merge(g_merged,m_merged,on=['Country Name', 'Year']),mPg_Pop,on=['Country Name', 'Year'])
merged_full
```

```python
#column renaming for title asthetics

merged_full.columns = ['Country_Name', 'Year', 'GDP_Abs(USD tril)', 'GDP_PerCapita(USD)', 'Mil_Exp(USD bil)', 'Mil_Exp_PerCapita(USD)', 'Share of GDP(%)', 'Population']
```

```python
merged_full
```

```python
#To sort(group) rows by country name and year both in ascending order

merged_full.sort_values(['Country_Name', 'Year'],ascending=[True,True],inplace=True)
merged_full
```
